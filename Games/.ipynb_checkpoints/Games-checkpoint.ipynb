## **Часть 1.** Изучение и подготовка датасета<a id='step_1'></a> 

Прежде чем проводить анализ данных, составление на их основе каких-либо выводов и проверки гипотез, необходимо осмотреть исходные данные, привести их к удобному для работы виду, заполнить, там где это возможно, пропуски. В первой части работы займемся именно этим.

### 1.1 Анализ исходных данных <a id='step_1_1'></a> 

В качестве исходных данных нам дана таблица `games.csv`. Выведем первые и последние её строки на экран, а также информацию о столбцах и изучим их.

```python
#Импорт всех библиотек, которые будут использованы в проекте
import pandas as pd
from IPython.display import display
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats as st
sns.set_style("darkgrid", {"axes.facecolor": ".9"})
sns.set(palette="muted")
```

```python
games = pd.read_csv('games.csv')
display(games.head(), games.tail())
display(games.info())
```

Предоставленная таблица с данными содержит сведения о 16715 играх, имеющих 11 параметров. Перечислю все параметры, опишу их и предложу что с ними стоит сделать, чтобы дальше работать с таблицой было удобнее:

- `Name` — название игры. Тип данных *object*, подходит для этого параметра. Имеется 2 строки с пропущенным названием, возможно из-за ошибки кодировки названия, или по какой-то иной причине. Необходимо будет избавиться от этих строк.
- `Platform` — платформа. Тип данных *object*, подходит для этого параметра. Пропусков в столбце нет.
- `Year_of_Release` — год выпуска. Тип данных *float64*, необходимо заменить на *int16*. Имеется около 270 пропусков.
- `Genre` — жанр игры. Тип данных *object*, подходит для этого параметра. Имеется 2 строки с пропусками, есть подозрение что это будут те же строки, что и пропуски в строках с названиями.
- `NA_sales` — продажи в Северной Америке (миллионы проданных копий). Тип данных *float64*, подходит для этого параметра. Пропусков нет.
- `EU_sales` — продажи в Европе (миллионы проданных копий). Тип данных *float64*, подходит для этого параметра. Пропусков нет.
- `JP_sales` — продажи в Японии (миллионы проданных копий). Тип данных *float64*, подходит для этого параметра. Пропусков нет.
- `Other_sales` — продажи в других странах (миллионы проданных копий). Тип данных *float64*, подходит для этого параметра. Пропусков нет.
- `Critic_Score` — оценка критиков (максимум 100). Тип данных *float64*, необходимо заменить на *int16*. В столбце имеется более 8,5 тыс. пропусков.
- `User_Score` — оценка пользователей (максимум 10). Тип данных *object*, необходимо заменить на *int16*. Имеется около 7 тыс. пропусков
- `Rating` — рейтинг от организации ESRB (англ. Entertainment Software Rating Board). Эта ассоциация определяет рейтинг компьютерных игр и присваивает им подходящую возрастную категорию. Тип данных *object*, подходит для этого параметра. Имеется около 7 тыс. пропусков.

Сразу удалим строки с пропусками в столбцах `Name` и `Genre`, поскольку игры без названия и жанра нам в исследовании не очень нужны. К тому же, потеряв 2 строки из 16715 мы не испортим статистику.

```python
games.dropna(subset = ['Name', 'Genre'], inplace = True) #удаление строчек, содержащих пропуски в столбцах `Name` и `Genre`
games.reset_index(drop = True, inplace = True)           #сброс индексов
games.info()
```

Предположение о том, что пропуски в обоих столбцах содержали одни и те же строчки оказалось верно. Избавившись от пропусков в обоих столбцах в таблице стало только на 2 строки меньше. Теперь можно приступать к дальней шей обработке данных. Следующим шагом изменим названия столбцов.

### 1.2 Изменение названий столбцов <a id='step_1_2'></a> 

Для удобства работы с таблицей в дальнейшем приведем названия всех столбцов в нижний регистр. 

```python
games.columns = games.columns.str.lower()   #перевод названий столбцов в нижний регистр
display(games.head(1))
```

Названия столбцов приведены к нижнему регистру. Следующим шагом преобразуем типы данных к более удобным.

### 1.3 Преобразование типов данных <a id='step_1_3'></a> 

В первом пункте мы упоминали какие столбцы к какому типу лучше привести, но перечислим их еще раз для более удобного вида:

- `year_of_release` - Int16
- `critic_score` - Int16
- `user_score` - Int16

Сделаем это одной функцией, но перед этим нужно столбец `user_score` преобразовать в числовой из строкового. Учтем что в этом типе нам могут попастся не тольцо цифры, но и буквенные значения, поэтому применим конструкцию *try-except*.

```python
try:
    games['user_score'] = pd.to_numeric(games['user_score'])     #Преобразование строки в число
    print('Преобразование проведено успешно')
except: 
    print('Ошибка в исходных данных')
```

Значит помимо значений, которые можно преобразовать в число, в строке содержатся и другие данные. Попробуем взглянуть на полный список оценок.

```python
games['user_score'].unique()
```

Смотрим внимательно на список и видим в середине четвертой строчки значение **tbd**. Это означает *to be announced*, т.е. оценка будет скоро объявлена. По сути, для нас нет разницы между этим значением и отсутствием оценки, поэтому заменим их на NaN и попробуем преобразовать в числа еще раз.

```python
games.loc[games['user_score'] == 'tbd', 'user_score'] = np.nan  #Замена всех значений 'tbd' в NaN
try:
    games['user_score'] = pd.to_numeric(games['user_score'])
    print('Преобразование проведено успешно')
except: 
    print('Ошибка в исходных данных')
```

Отлично, значит других аномалий в данных не было. Так же я считаю что необходимо привести оба вида оценки к единой шкале, и удобнее будет использовать шкалу 0-100, для чего `user_score` умножим на 10, после чего можно будет преобразовать все 3 столбца в Int16.

```python
games['user_score'] = games['user_score'] * 10
games = games.astype({"year_of_release":'Int16',        #Преобразование типов столбцов в Int16
                      "critic_score":'Int16',
                      "user_score":'Int16'})
games.info()
```

Все типы данных приведены к удобным для дальнейшего использования. Мы можем приступить к обработке пропусков.

### 1.4 Обработка пропусков <a id='step_1_4'></a> 

Как было указано в самом начале пропуски наблюдаются в четырех столбцах: *год выпуска игры, оценки критиков, оценки пользователей и ESRB рейтинг игры*.  Заполнять рейтинг ESRB может быть не совсем корректным - рейтингов у игр не так много, и если подсчитывать самый популярный ретинг, скажем, для жанра, и заполнить им пропуски, то это может привести нас в итоге к неверным выводам.  Поскольку пропусков в графах с оценками пользователей и критиков около 50%, то заполняя их некоторыми усредненными значениями мы также можем навредить дальнейшим выводам, например, влиянию оценок пользователей на продажи, поэтому заполнять пропуски в этих столбцах тоже не будем. Остается только столбец с пропусками года выпуска игры.

Поскольку мы имеем информацию о платформе, на которой вышла игра, а период "жизни" игровых платформ обычно ограничен, пропуски в графе с годом выхода игры мы можем заполнить медианными значениями года выпуска игр на каждой платформе.

```python
for platform, group_data in games.groupby('platform'):
    
    #рассчет медианного года выпуска игр для каждой платформы
    median_yor = round(group_data['year_of_release'].median()) 
    
    #Заполнение пропусков с помощью полученной медианы
    games.loc[(games['platform'] == platform) & (games['year_of_release'].isna()), 'year_of_release'] = median_yor 

print(games.info())
```

Пропуски в годах выхода игры заполнены, отлично. Можно переходить к следующему пункту.

Заполним пропуски в рейтинге ESRB значение ND - No Data

```python
games['rating'] = games['rating'].fillna('ND')
```

### 1.5 Подсчет суммарных продаж <a id='step_1_5'></a> 

На последнем этапе первичной подготовки подсчитаем суммарные продажи по всем регионам и запишем в отдельный столбец `all_sales`.

```python
#суммирование проданных копий по всем регионам для получения общего количества проданных копий
games['all_sales'] = games['na_sales'] + games['eu_sales'] + games['jp_sales'] + games['other_sales']   
display(games.head(), games.tail())
```

Суммарные продажи подсчитаны без проблем. 

### 1.6 Вывод <a id='step_1_6'></a> 

В этой части работы мы изучили предоставленные нам данные о продажах игр, привели названия и типы столбцов к более удобным для работы, заполнили пропуски в столбцах с оценками пользователей и критиков, а также посчитали для каждой игры суммарные продажи во всех регионах.

## **Часть 2.** Исследовательский анализ данных<a id='step_2'></a> 

Теперь, когда данные подготовлены, можно приступить к их исследованию. В этой части мы рассмотрим как выпускались игры по годам, изучим игровые платформы, определим актуальный на настоящее время период выпуска игр, изучим динамику продаж по платформам и распределение игр по жанрам.

### 2.1 Обзор выпука игр по годам <a id='step_2_1'></a> 

Для начала рассмотрим количество игр, выходивших в каждом году, для этого построим гистограмму распределения. Так мы сможем оценить период, в который выходило довольно мало игр, которые сейчас скорее всего покупать не будут.

```python
plt.figure(figsize = (9,6))
sns.distplot(games['year_of_release'], 
             bins=37, 
             hist_kws={'alpha':.9}, 
             kde=False)
plt.title('Распределение выпуска игр по годам')
plt.xlabel('Год')
plt.ylabel('Количество игр')
plt.show()
```

На графике видно что активное развиваться игровая индустрия стала в середине девяностых, начиная с 1994 года каждый год выходило более 100 популярных игр. Пик развития игровой индустрии пришелся на 2007-2008 года, и составлял более 1400 выпущеных в год игр после чего упал в ~2,5 раза к 2014-2016 годам, до значения около 500-600 игр.

### 2.2 Анализ игровых платформ <a id='step_2_2'></a> 

Далее, рассмотрим игровые платформы: найдем самые популярные платформы по продажам и изучим их "жизненный цикл", т.е. в каком году для них стали выпускать игры, которые были активно выкуплены и в каком прекратили. Для начала построим распределение количества проданных во всем мире копий по платформам

```python
#создание сводной таблицы по платформам
gpt = games.pivot_table(index = 'platform', 
                        values = 'all_sales', 
                        aggfunc = 'sum').sort_values('all_sales') 

gpt.plot(kind = 'bar', figsize = (9,6))                                                                     
plt.title('Проданные копии игр по платформам')
plt.xlabel('Платформа')
plt.ylabel('Продано копий игр, млн.')
plt.show()   
```

На распределении можно выделить 6 наиболее популярных платформ, для каждой из которых было продано более 600 млн. копий игр: PS, DS, Wii, PS3, X360, PS2. Построим для них распределение продаж игр, выпущенных в каждом году.

```python
#формируем сводную таблицу, отсортированную по платформам и годам выпуска игр, где для каждой пары посчитаем сумму всех проданных копий игр
games_by_platform_year = games.pivot_table(index = ['platform', 'year_of_release'], values = 'all_sales', aggfunc = 'sum').reset_index()

#определяем из прошлой таблицы список платформ, сумма продаж копий всех игр которых превышает 600 млн.
top_sales_platform = gpt.reset_index().query('all_sales > 600')['platform'].unique()

#Для каждой платформы строим график зависимости количества проданных копий от года выпуска
plt.figure(figsize = (11,8))
for platform in top_sales_platform:
    sns.lineplot(x = "year_of_release", 
                 y = "all_sales", 
                 label = platform, 
                 data = games_by_platform_year.query('platform == @platform and year_of_release > 1990'), 
                 linewidth = 3)
plt.title('Проданные копии игр популярных платформ')
plt.xlabel('Год выпуска')
plt.ylabel('Продано копий игр, млн.')
plt.show()
```

На графике выше виден жизненный цикл платформ, составляющий около 8-10 лет. А также можно сделать вывод о том, что самые популярные платформы к 2016 году уже либо завершили свой жизненный цикл, либо на самом его исходе. Это наводит на мысль о том, что перспективные на данный момент платформы еще не успели продать столько копий игр, чтобы сравняться с самыми популярными платформами.

Основываясь на изложенных выше данных мы можем определить актуальный период, на который стоит обращать внимание при планировании продаж на 2017 год. Я полагаю что стоит выбрать половину жизненного цикла платформ - 5 лет, т.е. начиная с 2012 года.

### 2.3 Анализ динамики продаж по платформам <a id='step_2_3'></a> 

В предыдущем пункте мы определили актуальный период. Теперь рассмотрим все игры, выпущенные в этот период и определим перспективные платформы, жизненный цикл которых только начался. Для этого выделим таблицу игр, выпущенных в 2012 году и после (`games_new`), и на её основе построим для каждой платформы график зависимости количества проданных копий игр от года их выхода.

```python
plt.figure(figsize = (11,8))

#Выделение данных об играх, выпущенных в 2012 году и позже
games_new = games.query('year_of_release >= 2012')

#Формирование сводной таблицы из новых данных
games_new_by_platform_year = games_new.pivot_table(index = ['platform', 'year_of_release'], 
                                                   values = 'all_sales', 
                                                   aggfunc = 'sum').reset_index() 

#Построение в цикле графиков для каждой платформы
for platform in games_new['platform'].unique():
    sns.lineplot(x = "year_of_release", 
                 y = "all_sales", 
                 label = platform, 
                 data = games_new_by_platform_year.query('platform == @platform'), 
                 linewidth = 3)
plt.title('Проданные копии игр в актуальный период')
plt.xlabel('Год выпуска')
plt.ylabel('Продано копий игр, млн.')
plt.show()    
```

Много линий. Давайте разбираться. Во-первых нужно разу сказать что не стоит обращать внимание на то, что игры выпущенные в 2016 году были проданы в меньшем количестве, чем игры выпущенные в 2014-2015 годах, т.к. скорее всего это связанно с тем, что их еще просто не успели продать в достаточном количестве. На графике можно выделить 4 платформы, которые появились менее 8 лет назад: *3DS, PSV, XOne, PS4*. Причем последние 2 появились позже и имеют большее количество проданных копий. Все остальные платформы, попавшие на график находятся в конце своего жизненного цикла, но основываясь на данных 2014-2015 года есть еще небольшая доля вероятности того, что игры произведенные для *PS3* и *X360* еще будут продаваться в 2017 году. 

Конечно, нам бы еще лучше помогла динамика продаж каждой игры, т.е. сколько копий каждой игры было продано в каком году, но поскольку мы ею не обладаем, то опираемся на предположения. Также, хоть на графике эта линия особо не выделяется, но есть одна платформа, которая еще несколько лет точно будет актуальна, хоть её продажи на фоне *PS4* и *XOne* не так велики - это *PC*, и отбрасывать её было бы ошибкой.

Суммируя вышесказанное я выделяю несколько платформ, на которые стоит обратить внимание: *PS4, XOne, PSV, 3DS, PS3, X360, PC.*

Далее посмотрим динамику продаж внутри этих платформ, построим диаграммы размаха для каждой платформы по продажам.

### 3.1 Самые популярные платформы по регионам <a id='step_3_1'></a> 

Подсчитывать популярные платформы по регионам просто, нужно просто посчитать сумму продаж игр по каждой платформе отдельно для каждого региона, а потом, для удобства прочтения, соберем эти данные в единую таблицу.

```python
#Функция подсчета всех параметров, не вошедших в топ-5, для отрисовки правильного пайчарта
def table_for_pie(x, cntry, ind):
    su = 0
    for i in x.index:
        if i not in x.head().index:
            su += x.loc[i, cntry]
    row = pd.Series(['Other', su],index=[ind,cntry])
    x = x.head().reset_index().append(row, ignore_index = True).set_index(ind)
    return x

#Функция для вывода на пайчарте "абсолютных(относительных %) значений"
def make_autopct(values):
    def my_autopct(pct):
        total = sum(values)
        val = int(round(pct*total/100.0))
        return '{v:d} ({p:.0f}%)'.format(p=pct,v=val)
    return my_autopct

#Функция, раскрашивающая поочередно столбцы в таблице
def highlight(x):
    c1 = 'background-color: beige'
    c2 = 'background-color: bisque' 
    df1 = pd.DataFrame(c2, index=x.index, columns=x.columns)
    for i in range(len(x.columns)):
        if i % 2 == 1:
            df1.loc[:, x.columns[i]] = c1
    return df1
```

```python
#перечень названий столбцов с продажами в регионах
countries = {'na_sales': "Северной америке", 'eu_sales': "Европе", 'jp_sales': "Японии"}

#формирование пустой таблицы, в которую будут собираться топ-5 платформ
top_5 = {'platform':[0, 1, 2, 3, 4]}                      
top_5_platform = pd.DataFrame(data = top_5)

#Подготовка площадок для вывода пайчартов
fig, (bx0, bx1) = plt.subplots(nrows=1, ncols=2, figsize=(14, 7))
fig1, bx2 = plt.subplots(figsize=(14, 7))

#перечень названий столбцов с продажами в регионах
bx = {'na_sales': bx0, 'eu_sales': bx1, 'jp_sales': bx2}        

#Таблицу для вывода на экран данных о топ-5 плтформах делаем так:
#Считаем для каждого региона данные о топ 5 платформах и собраем это в отдельную таблицу, 
#поочередно "приклеивая" новые данные справа
for country in countries:
    cntry_games = games.pivot_table(index = 'platform', 
                                    values = country, 
                                    aggfunc = 'sum').sort_values(by = country, ascending = False)
    
    top_5_platform = top_5_platform.join(cntry_games.head().reset_index(), lsuffix = 'a', rsuffix = 'b')
    games_pie = table_for_pie(cntry_games, country, 'platform')
    bx.get(country).pie(games_pie[country], 
                        labels=games_pie.index, 
                        autopct = make_autopct(games_pie[country]),
                        shadow=True, 
                        startangle=90)
    bx.get(country).axis('equal')
    title = 'Продажи по платформам в ' + countries.get(country)
    bx.get(country).set_title(title)

#Переименовываем столбцы    
top_5_platform.columns = ['offtop', 'na_top', 'na_sales', 'eu_top', 'eu_sales', 'jp_top', 'jp_sales'] 
#Избавляемся от бессмысленного начального столбца
top_5_platform.drop(columns = 'offtop', inplace = True)                                               
plt.show()

display(top_5_platform.style.apply(highlight, axis = None))
```

Вот и наша табличка, в ней видно пьедестал платформ, и сколько копий игр для каждой из топ-5 платформ было продано в каждом регионе.

Рынки Северной Америки и Европы похожи, в каждом из них есть *X360, PS2, PS3* и *Wii*, различия только на пятом месте в *DS* (NA) и *PS*(EU). В Европе самая популярная *PS2*, в Америке -  *X360*. Довольно сильно отличается от них Япония, тут 3 из 5 платформ - консоли японской фирмы Nintendo: DS, 3DS, SNES, и оставшиеся две от японской фирмы Playstation: *PS* и *PS2*. Причем почти все эти платформы, кроме *3DS*, более старые, чем популярные в Америке и Европе *X360* и *PS3*. 

Также видна разница размеров этих трех рынков - большинство продаж в Северной Америке, меньше в Европе и самое маленькое в Японии. Полагаем, что это связано с населением и культурными особенностями: в Японии проживает меньше людей, чем в Северной Америке или Европе. Но несмотря на то, что в Европе население больше Америки, по-видимому в последней более развит рынок компьютерных игр и в них играет больше людей.

```python
top_5_genre = pd.DataFrame(data = top_5)
#таблицу для вывода топ-5 жанров в разных регионах формируем так же, как это сделали для топ-5 платформ.
fig, (cx0, cx1) = plt.subplots(nrows=1, ncols=2, figsize=(14, 7))
fig1, cx2 = plt.subplots(figsize=(14, 7))

#перечень названий столбцов с продажами в регионах
cx = {'na_sales': cx0, 'eu_sales': cx1, 'jp_sales': cx2}        

for country in countries:
    cntry_games = games.pivot_table(index = 'genre', 
                                    values = country, 
                                    aggfunc = 'sum').sort_values(by = country, ascending = False)
    
    top_5_genre = top_5_genre.join(cntry_games.head().reset_index(), lsuffix = 'a', rsuffix = 'b')
    games_pie = table_for_pie(cntry_games, country, 'genre')
    
    cx.get(country).pie(games_pie[country], 
                        labels=games_pie.index, 
                        autopct = make_autopct(games_pie[country]),
                        shadow=True, 
                        startangle=90)
    cx.get(country).axis('equal')
    title = 'Продажи по жанрам в ' + countries.get(country)
    cx.get(country).set_title(title)
plt.show()    
top_5_genre.columns = ['offtop', 'na_top', 'na_sales', 'eu_top', 'eu_sales', 'jp_top', 'jp_sales']
top_5_genre.drop(columns = 'offtop', inplace = True)
display(top_5_genre.style.apply(highlight, axis = None))
```

Полагаем, что это связано с культурными особенностями, а так же с тем, что на Японские платформы было выпущено гораздо больше игр жанра *RPG*.

## **Часть 4.** Проверка гипотез<a id='step_4'></a> 

В этой части работы мы будем проверять гипотезы, связанные с пользовательскими оценками игр. О том, что рейтинги игр на платформах *Xbox One* и *PC* одинаковые, а также о том, что рейтинги жанров *Action* и *Sports* разные.

### 4.1 Средние пользовательские рейтинги платформ Xbox One и PC одинаковые <a id='step_4_1'></a> 

Выпишем выборки, необходимые нам в работе в следующие переменные:

- `s_xbox_us` - оценки пользователей для игр платформы Xbox One
- `s_pc_us` - оценки пользователей для игр платформы PC
- `s_action_us` - оценки пользователей для игр жанра Action
- `s_sports_us` - оценки пользователей для игр жанра Sports

А также посчитаем для каждой из этих выборок значение дисперсий.

```python
s_xbox_us = games[games['platform'] == 'XOne']['user_score'].dropna()
s_pc_us   = games[games['platform'] == 'PC']['user_score'].dropna()
s_action_us = games[games['genre'] == 'Action']['user_score'].dropna()
s_sports_us = games[games['genre'] == 'Sports']['user_score'].dropna()

#Поскольку выше мы посчитали, что дисперсии пользовательских оценок игр, выпущенных на этих платформах, различаются я задам ddof = 1
print('Дисперсии пользовательских оценок для:')
print('XBox One:', round(np.var(s_xbox_us, ddof = 1), 2))
print('PC:', round(np.var(s_pc_us, ddof = 1), 2))
print('Action:', round(np.var(s_action_us, ddof = 1), 2))
print('Sports:', round(np.var(s_sports_us, ddof = 1), 2))
```

Построим графики распределения оценок во всех четырех категориях для оценки характера распределения.

```python
f, axes=plt.subplots(2, 2, figsize=(14, 12))

sns.distplot(s_xbox_us, bins=50, color="orange", hist_kws={'alpha':.9}, kde=False, ax=axes[0, 0])
axes[0,0].plot()
axes[0,0].set_title('XOne')
axes[0,0].set_xlabel('Оценка')
axes[0,0].set_ylabel('Количество оценок')

sns.distplot(s_pc_us, bins=50, color="dodgerblue", hist_kws={'alpha':.6}, kde=False, ax=axes[0, 1])
axes[0,1].plot()
axes[0,1].set_title('PC')
axes[0,1].set_xlabel('Оценка')
axes[0,1].set_ylabel('Количество оценок')
                 
sns.distplot(s_action_us, bins=50, color="green", hist_kws={'alpha':.9}, kde=False, ax=axes[1, 0])
axes[1,0].plot()
axes[1,0].set_title('Action')
axes[1,0].set_xlabel('Оценка')
axes[1,0].set_ylabel('Количество оценок')

sns.distplot(s_sports_us, bins=50, color="red", hist_kws={'alpha':.9}, kde=False, ax=axes[1, 1])
axes[1,1].plot()
axes[1,1].set_title('Sports')
axes[1,1].set_xlabel('Оценка')
axes[1,1].set_ylabel('Количество оценок')
```

Во всех четырех категориях данные не подчиняются закону нормального распределения. Поскольку оба раза мы будем сравнивать две количественные, ненормально распределенные, несвязанные группы данных, поэтому будем использовать **критерий Манна-Уитни**.

Теперь можно проверять гипотезу. Критический уровень статистической значимости возьмем равным 5%. Сформулируем гипотезу следующим образом:

- Средние пользовательские рейтинги игр платформ Xbox One и PC **равны**

```python
alpha = .05 # критический уровень статистической значимости

#Поскольку выше мы посчитали, что дисперсии пользовательских оценок игр, выпущенных на этих платформах, различаются зададим equal_var = False
results = st.mannwhitneyu(s_xbox_us, s_pc_us, alternative = 'two-sided')
                                                                      
print('p-значение:', results.pvalue)

if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

Вероятность того, что средние пользовательские рейтинги игр платформ равна 1*10^(-8)% что, безусловно, дает нам полные основания отвергнуть эту гипотезу.

### 4.2 Средние пользовательские рейтинги игр жанров Action и Sports разные <a id='step_4_2'></a> 

Теперь проверим гипотезу о том, что средние пользовательские рейтинги игр жанров Action и Sports отличаются. Но проверять мы будем обратную, нулевую, гипотезу, а на её основании сделаем вывод об этой. Нулевую гипотезу сформулируем следующим образом:

- Средние пользовательские рейтинги игр жанров Action и Sports **равны**

А также сформулируем альтернативную гипотезу, которая нас и интересует:

- Средние пользовательские рейтинги игр жанров Action и Sports **не равны**

```python
#Поскольку выше мы посчитали, что дисперсии пользовательских оценок игр, выпущенных на этих платформах, различаются зададим equal_var = False
results2 = st.mannwhitneyu(s_action_us, s_sports_us, alternative = 'two-sided')
                                                                      
print('p-значение:', results2.pvalue)

if (results2.pvalue < alpha):
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

Нулевая гипотеза не отвергнута, а если применить к нашей изначальной, можено сказать так: предположив, что пользовательские рейтинги игр жанров Action и Sports разные, мы ошибемся в 71,8% случаев, что для нас неприемлимо. То есть гипотеза не подтверждена.

### 4.3 Вывод <a id='step_4_3'></a> 

В данной части мы проверили две гипотезы и можем сказать что:

1. Гипотеза о том, что средние пользовательские рейтинги игр платформ Xbox One и PC одинаковы **отвергнута**.
2. Гипотеза о том, что средние пользовательские рейтинги игр жанров Action и Sports разные **не подтверждена**.
